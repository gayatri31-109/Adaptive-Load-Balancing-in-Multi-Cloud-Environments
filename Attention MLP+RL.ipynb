# Attention MLP

```python
# =========================================
# 1. IMPORT LIBRARIES
# =========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import tensorflow as tf
from tensorflow.keras.layers import Dense, Input, Multiply, Softmax
from tensorflow.keras.layers import BatchNormalization, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau


# =========================================
# 2. LOAD DATA
# =========================================

path = r"C:\Users\hp\Desktop\Major Pr\borg_traces_data.csv"

df = pd.read_csv(path)

print("Dataset shape:", df.shape)


# =========================================
# 3. DROP USELESS COLUMNS
# =========================================

drop_cols = [
'Unnamed: 0',
'constraint',
'user',
'collection_name',
'collection_logical_name',
'start_after_collection_ids',
'cpu_usage_distribution',
'tail_cpu_usage_distribution'
]

df.drop(columns=drop_cols, inplace=True, errors='ignore')


# =========================================
# 4. FEATURE ENGINEERING
# =========================================

df['runtime'] = df['end_time'] - df['start_time']
df['delay'] = df['start_time'] - df['time']


# =========================================
# 5. HANDLE MISSING VALUES
# =========================================

df.fillna({
'vertical_scaling':0,
'scheduler':0,
'cycles_per_instruction':df['cycles_per_instruction'].median(),
'memory_accesses_per_instruction':df['memory_accesses_per_instruction'].median()
}, inplace=True)

df.fillna(0, inplace=True)


# =========================================
# 6. ENCODE EVENT
# =========================================

df['event'] = df['event'].astype('category').cat.codes


# =========================================
# 7. KEEP ONLY NUMERIC
# =========================================

df = df.select_dtypes(include=[np.number])


# =========================================
# 8. SELECT FEATURES
# =========================================

features = [
'priority',
'machine_id',
'assigned_memory',
'page_cache_memory',
'cycles_per_instruction',
'memory_accesses_per_instruction',
'runtime',
'delay',
'event',
'failed'
]

targets = [
'assigned_memory',
'page_cache_memory',
'runtime'
]

X = df[features]
y = df[targets]


# =========================================
# 9. FIX INF VALUES
# =========================================

X.replace([np.inf,-np.inf],np.nan,inplace=True)
X.fillna(0,inplace=True)


# =========================================
# 10. SCALE FEATURES
# =========================================

X_scaler = StandardScaler()
X_scaled = X_scaler.fit_transform(X)


# =========================================
# 11. SCALE TARGETS
# =========================================

y_scaler = StandardScaler()
y_scaled = y_scaler.fit_transform(y)


# =========================================
# 12. TRAIN TEST SPLIT
# =========================================

X_train,X_test,y_train,y_test = train_test_split(
X_scaled,
y_scaled,
test_size=0.2,
random_state=42
)


# =========================================
# 13. ATTENTION MLP MODEL
# =========================================

input_layer = Input(shape=(X_train.shape[1],))


# Attention weights
attention = Dense(X_train.shape[1],activation='tanh')(input_layer)
attention = Softmax()(attention)

# Apply attention
attended_features = Multiply()([input_layer,attention])


# MLP Layers
x = Dense(256,activation='relu')(attended_features)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = Dense(128,activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = Dense(64,activation='relu')(x)

output = Dense(3)(x)

model = Model(inputs=input_layer,outputs=output)


model.compile(
optimizer=tf.keras.optimizers.Adam(0.001),
loss='mse',
metrics=['mae']
)

model.summary()


# =========================================
# 14. CALLBACKS
# =========================================

early_stop = EarlyStopping(
monitor='val_loss',
patience=6,
restore_best_weights=True
)

reduce_lr = ReduceLROnPlateau(
monitor='val_loss',
factor=0.3,
patience=3
)


# =========================================
# 15. TRAIN MODEL
# =========================================

history = model.fit(
X_train,
y_train,
epochs=50,
batch_size=256,
validation_split=0.2,
callbacks=[early_stop,reduce_lr],
verbose=1
)


# =========================================
# 16. PREDICTIONS
# =========================================

y_pred_scaled = model.predict(X_test)

y_pred = y_scaler.inverse_transform(y_pred_scaled)
y_test_real = y_scaler.inverse_transform(y_test)


# =========================================
# 17. METRICS
# =========================================

print("MAE:",mean_absolute_error(y_test_real,y_pred))
print("RMSE:",np.sqrt(mean_squared_error(y_test_real,y_pred)))
print("R2:",r2_score(y_test_real,y_pred))


# =========================================
# 18. TRAINING CURVE
# =========================================

plt.figure(figsize=(8,4))
plt.plot(history.history['loss'],label='Train Loss')
plt.plot(history.history['val_loss'],label='Validation Loss')
plt.legend()
plt.title("Training vs Validation Loss")
plt.show()


# =========================================
# 19. CPU PREDICTION
# =========================================

plt.figure(figsize=(8,4))
plt.plot(y_test_real[:500,0],label='Actual CPU')
plt.plot(y_pred[:500,0],label='Predicted CPU')
plt.legend()
plt.title("CPU Prediction")
plt.show()


# =========================================
# 20. MEMORY PREDICTION
# =========================================

plt.figure(figsize=(8,4))
plt.plot(y_test_real[:500,1],label='Actual Memory')
plt.plot(y_pred[:500,1],label='Predicted Memory')
plt.legend()
plt.title("Memory Prediction")
plt.show()


# =========================================
# 21. BANDWIDTH PREDICTION
# =========================================

plt.figure(figsize=(8,4))
plt.plot(y_test_real[:500,2],label='Actual Bandwidth')
plt.plot(y_pred[:500,2],label='Predicted Bandwidth')
plt.legend()
plt.title("Bandwidth Prediction")
plt.show()
```

```python
import numpy as np
import gymnasium as gym
from gymnasium import spaces
import matplotlib.pyplot as plt

from stable_baselines3 import DQN, PPO
```

```python
class CloudEnv(gym.Env):

    def __init__(self, X, y):

        super(CloudEnv, self).__init__()

        self.X = X
        self.y = y
        self.index = 0

        # STATE SPACE
        self.observation_space = spaces.Box(
            low=-np.inf,
            high=np.inf,
            shape=(X.shape[1],),
            dtype=np.float32
        )

        # ACTION SPACE
        # 0 → Cloud A
        # 1 → Cloud B
        # 2 → Cloud C
        self.action_space = spaces.Discrete(3)

    def reset(self, seed=None, options=None):

        self.index = 0
        state = self.X[self.index]

        return state, {}

    def step(self, action):

        cpu = self.y[self.index][0]
        memory = self.y[self.index][1]
        runtime = self.y[self.index][2]

        reward = 0

        # Reward design for balanced load
        if cpu < 0.7:
            reward += 1
        else:
            reward -= 1

        if memory < 0.7:
            reward += 1
        else:
            reward -= 1

        if runtime < 1000:
            reward += 1

        self.index += 1

        done = self.index >= len(self.X)

        if done:
            next_state = self.X[-1]
        else:
            next_state = self.X[self.index]

        return next_state, reward, done, False, {}
```

```python
env = CloudEnv(X_scaled, y_scaled)
```

```python
model_dqn = DQN(
    "MlpPolicy",
    env,
    learning_rate=0.0003,
    buffer_size=50000,
    learning_starts=1000,
    batch_size=64,
    verbose=1
)

model_dqn.learn(total_timesteps=10000)
```

# DQN AGENT TRAININIG

```python
obs, _ = env.reset()

total_reward = 0

for i in range(200):

    action, _ = model_dqn.predict(obs)

    obs, reward, done, _, _ = env.step(action)

    total_reward += reward

    print("Step:", i,
          "Action:", action,
          "Reward:", reward)

    if done:
        break

print("Total Reward:", total_reward)
```

# PPQ

```python
from stable_baselines3 import PPO

model_ppo = PPO(
    "MlpPolicy",
    env,
    learning_rate=0.0003,
    n_steps=2048,
    batch_size=64,
    verbose=1
)

model_ppo.learn(total_timesteps=10000)
```

```python
ppo_rewards = []

obs, _ = env.reset()

for i in range(500):

    action, _ = model_ppo.predict(obs)

    obs, reward, done, _, _ = env.step(action)

    ppo_rewards.append(reward)

    if done:
        break


plt.figure(figsize=(8,4))

plt.plot(ppo_rewards)

plt.xlabel("Steps")
plt.ylabel("Reward")
plt.title("PPO Load Balancing Reward")

plt.show()
```

# COMPARE DQN & PPQ

```python
dqn_rewards = []
ppo_rewards = []

obs, _ = env.reset()

for i in range(200):

    action, _ = model_dqn.predict(obs)
    obs, reward, done, _, _ = env.step(action)
    dqn_rewards.append(reward)

    if done:
        break

obs, _ = env.reset()

for i in range(200):

    action, _ = model_ppo.predict(obs)
    obs, reward, done, _, _ = env.step(action)
    ppo_rewards.append(reward)

    if done:
        break

plt.figure(figsize=(8,4))

plt.plot(dqn_rewards,label="DQN")
plt.plot(ppo_rewards,label="PPO")

plt.xlabel("Steps")
plt.ylabel("Reward")
plt.title("DQN vs PPO Performance")

plt.legend()

plt.show()
```
